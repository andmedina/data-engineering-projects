RandomUser is a free API that provides randomly generated users for testing. 
It can return multiple users with details like gender, email, image, username, address, name, and more.

Fruityvice API provides data and interesting information on various fruits.


```python
from randomuser import RandomUser
import pandas as pd
```

```python
#Create a random user object
r = RandomUser()
```

```python
#Then, using generate_users() function, we get a list of random 10 users.
some_list = r.generate_users(10)
```

```python
some_list
```

```python
#Let's say we only need 10 users with full names and their email addresses. We can write a "for-loop" to print these 10 users.
name = r.get_full_name()

for user in some_list:
    print (user.get_full_name()," ",user.get_email())
```

```python
#Generate photos of 10 users
for user in some_list:
    print(user.get_picture())
```

```python
# Generate a table with user info using a function with desired parameters (e.g., name, gender, city).
# Parameters depend on test requirements. Use Get Methods from the start of this notebook.
# Return a pandas dataframe with the users.
#Documntation found here: https://randomuser.me/documentation

def get_users():
    users =[]
     
    for user in RandomUser.generate_users(10):
        users.append({"Name":user.get_full_name(),"Gender":user.get_gender(),"City":user.get_city(),"State":user.get_state(),"Email":user.get_email(), "DOB":user.get_dob(),"Picture":user.get_picture()})
      
    return pd.DataFrame(users)   
```

```python
get_users()
```

```python
#If we run the function again it will create another 10 different users. 
get_users()
```

```python
#We can also have a dataframe which we can use for further test cases.
df1 = pd.DataFrame(get_users())
```

Fruityvice API
We will obtain the fruityvice API data using requests.get("url") function. The data is in a json format.

```python
import requests
import json
```

```python
data = requests.get("https://fruityvice.com/api/fruit/all")
```

```python
results = json.loads(data.text)
```

```python
pd.DataFrame(results)
```

```python
df2 = pd.json_normalize(results)
```

```python
df2
```

```python
cherry = df2.loc[df2["name"] == 'Cherry']
(cherry.iloc[0]['family']) , (cherry.iloc[0]['genus'])
```

```python
#find out how many calories are contained in a banana.
cal_banana = df2.loc[df2["name"] == 'Banana']
cal_banana.iloc[0]['nutritions.calories']
```
